```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import tempfile
import netCDF4 as nc
from keras.models import load_model
```

```python
def load_your_model(model_path):
    model = load_model(model_path)
    return model
model = load_your_model('/home/jovyan/shared/jiarui-notebooks/CNNLSTM_2002_2017.keras')
```

```python
dates = ["20230812", "20230813", "20230814", "20230815", "20230816"]
base_path = "/home/jovyan/shared/jiarui-notebooks/shared/"

datasets = []
for date in dates:
    file_path = base_path + date + ".nc"
    ds = xr.open_dataset(file_path)
    datasets.append(ds)
```

```python
def preprocess_vis_input_data(day_data):
    day_data = np.squeeze(day_data)
    mean_val = np.nanmean(day_data)
    processed_data = day_data - mean_val
    # Replace NaNs with 0.0
    processed_data = np.where(np.isnan(processed_data), 0.0, processed_data)
    return processed_data

def postprocess_prediction(prediction, input_data_day):
    # Create land mask from the input day
    land_mask = create_land_mask(input_data_day)
    
    # Ensure dimensions match by squeezing the prediction
    prediction_squeezed = np.squeeze(prediction)

    # Set those positions in the prediction to NaN
    prediction_squeezed[land_mask] = np.nan

    # Add back the historical mean
    mean_val = np.nanmean(input_data_day)
    prediction_post = np.where(np.isnan(prediction_squeezed), np.nan, prediction_squeezed + mean_val)

    return prediction_post


def create_land_mask(data):
    data_copy = data.copy()
    data_copy[data_copy == 0] = np.nan
    return np.isnan(data_copy)
```

```python
preprocessed_data = [preprocess_vis_input_data(day['analysed_sst'].values) for day in datasets]
input_data = np.array(preprocessed_data)
```
